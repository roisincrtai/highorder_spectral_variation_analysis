# Jacobian

```python
def vec(x, column_major=True):
    if column_major:
        return x.t().flatten()
    else:
        return x.flatten()

def unvec(x, shape, column_major=True):
    if column_major:
        return x.unflatten(dim=0, sizes=shape[::-1]).t()
    else:
        return x.unflatten(dim=0, sizes=shape)

def singular_value_jacobian(U, S, Vh, k):
    sigma_k = S[k - 1]  # Largest singular value
    # Get the corresponding left and right singular vectors
    u_k = U[:, k - 1] 
    v_k = Vh[k - 1, :] 
    return torch.outer(u_k, v_k)
```

```python
# Set seed for reproducibility
torch.manual_seed(1)

# Define dimensions
m = 64  # Number of rows
n = 32  # Number of columns
k = 1   # The k-th largest singular value.

theta_mat = torch.randn(m, n, dtype=torch.double, requires_grad=True)

U, S, Vh = torch.linalg.svd(theta_mat.detach(), full_matrices=True)

sigma_k = S[k - 1]  # Largest singular value
# Get the corresponding left and right singular vectors
u_k = U[:, k - 1] 
v_k = Vh[k - 1, :] 

print("theta_mat.shape: ", theta_mat.shape)


def sigma_func(theta_mat):
    U, S, Vh = torch.linalg.svd(theta_mat, full_matrices=True)
    return S[k - 1]

y = sigma_func(theta_mat)

# Compute gradient using autograd.grad
J_autograd = torch.autograd.grad(y, theta_mat, create_graph=True)[0] 
print("J_autograd.shape: ", J_autograd.shape)

#J_theoretical = torch.outer(u_k, v_k)
J_theoretical = singular_value_jacobian(U, S, Vh, k=1)

print("J_theoretical.shape: ", J_theoretical.shape)

J_error = torch.norm(J_theoretical - J_autograd)

print()
print("J_autograd: ", J_autograd)
print("J_theoretical: ", J_theoretical)
print(f"error: {J_error.item()}")




#figsize = (4.5, 4.5)
pad_inches = 0.04

save_dir="plots"
os.makedirs(save_dir, exist_ok=True)

#plt.figure(figsize=figsize)
plt.figure()
plt.imshow(J_theoretical.numpy(), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "jacobian_theoretical.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('jacobian_theoretical')
plt.show()
print(f"Saved to {plot_filename}")

plt.figure()
plt.imshow(J_autograd.detach().numpy(), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "jacobian_numerical.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('jacobian_numerical')
plt.show()
print(f"Saved to {plot_filename}")


plt.figure()
plt.imshow(np.abs(J_theoretical.numpy() - J_autograd.detach().numpy()), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "jacobian_error.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('jacobian_error')
plt.show()
print(f"Saved to {plot_filename}")
```

# Hessian

```python
# Set seed for reproducibility
torch.manual_seed(1)

def vec(x, column_major=True):
    if column_major:
        return x.t().flatten()
    else:
        return x.flatten()

def unvec(x, shape, column_major=True):
    if column_major:
        return x.unflatten(dim=0, sizes=shape[::-1]).t()
    else:
        return x.unflatten(dim=0, sizes=shape)

def singular_value_hessian_full(U, S, Vh, k):
    sigma_k = S[k - 1].item()
    # Get the singular vectors
    u_k = U[:, k - 1] 
    v_k = Vh[k - 1, :] 
    
    m = U.size(0)
    n = Vh.size(0)
    
    H = torch.zeros((m*n, m*n), dtype=torch.float64)

    #left bases
    for i in range(m):
        if i == k - 1:
            continue
        
        if i >= len(S):
            sigma_i = 0
        else:
            sigma_i = S[i].item()

        coeff = sigma_k / (sigma_k**2 - sigma_i**2)

        v_k = Vh[k - 1, :]  
        u_i = U[:, i]  

        kron_vk_ui = torch.kron(v_k, u_i)  
    
        H += coeff * torch.outer(kron_vk_ui, kron_vk_ui) 

    #right bases
    for j in range(n):
        if j == k - 1:
            continue
        
        if j >= len(S):
            sigma_j = 0
        else:
            sigma_j = S[j].item()

        coeff = sigma_k / (sigma_k**2 - sigma_j**2)

        v_j = Vh[j, :]  
        u_k = U[:, k - 1]  

        kron_vj_uk = torch.kron(v_j, u_k)  
    
        H += coeff * torch.outer(kron_vj_uk, kron_vj_uk) 


    # cross terms.
    for l in range(len(S)):
        if l == k - 1:
            continue
        
        sigma_l = S[l].item()
        u_l = U[:, l]  
        v_l = Vh[l, :]  

        coeff = sigma_l / (sigma_k**2 - sigma_l**2)


        # Compute the Kronecker products
        kron_vk_ul = torch.kron(v_k, u_l)      
        kron_vl_uk = torch.kron(v_l, u_k)  

     
        # Add to the theoretical Hessian
        H += coeff * torch.outer(kron_vk_ul, kron_vl_uk) 
        H += coeff * torch.outer(kron_vl_uk, kron_vk_ul)
    
    return H






# Define dimensions
m = 10  # Number of rows
n = 5 # Number of columns
k = 1

# Initialize theta_vec with well-separated singular values
theta_mat = torch.randn(m, n, dtype=torch.float64)

print("theta_mat.shape: ", theta_mat.shape)

theta_vec = vec(theta_mat).detach()

theta_vec.requires_grad = True

    
def sigma_func(theta_vec):
    theta_mat = unvec(theta_vec, shape=(m, n))
    U, S, Vh = torch.linalg.svd(theta_mat, full_matrices=True)
    return S[k - 1]


# Compute numerical Hessian using automatic differentiation
H_autograd = torch.autograd.functional.hessian(sigma_func, theta_vec).detach()

# Reshape theta_vec into matrix form
theta_mat = unvec(theta_vec.detach(), shape=(m, n)).detach()

# Perform SVD on theta_mat
U, S, Vh = torch.linalg.svd(theta_mat.detach(), full_matrices=True)


print("theta_mat.shape: ", theta_mat.shape)
print("U.shape: ", U.shape)
print("len(S): ", len(S))
print("Vh.shape: ", Vh.shape)

#sigma_k = S[k - 1].item()
# Get the singular vectors
#u_k = U[:, k - 1] 
#v_k = Vh[k - 1, :] 

# Initialize the theoretical Hessian
H_theoretical = singular_value_hessian_full(U, S, Vh, k = k)


error = torch.norm(H_autograd - H_theoretical)
print(f"error: {error.item()}")

# Compute cosine similarity and ratio of norms
cosine = torch.nn.functional.cosine_similarity(H_theoretical.reshape(1, -1), H_autograd.reshape(1, -1))
ratio = torch.norm(H_theoretical) / torch.norm(H_autograd)
print(f"Cosine similarity: {cosine.item()}")
print(f"Ratio of norms: {ratio.item()}")

save_dir="plots"
os.makedirs(save_dir, exist_ok=True)

pad_inches = 0.04

# Plot the Hessians for visual comparison
plt.figure()
plt.imshow(H_autograd.numpy(), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "hessian_numerical.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('hessian_numerical')
plt.show()
print(f"Saved to {plot_filename}")

plt.figure()
plt.imshow(H_theoretical.numpy(), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "hessian_theoretical.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('hessian_theoretical')
plt.show()
print(f"Saved to {plot_filename}")

plt.figure()
plt.title("Error (H_theoretical - H_autograd)")
plt.imshow((H_theoretical - H_autograd).numpy(), aspect='auto')
plt.colorbar()
plot_filename = os.path.join(save_dir, "hessian_error.pdf")
plt.savefig(plot_filename, dpi = 600, format = 'pdf', bbox_inches='tight', pad_inches=pad_inches)
plt.title('hessian_error')
plt.show()
print(f"Saved to {plot_filename}")


plt.show()
```
